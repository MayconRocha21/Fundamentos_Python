### Importando o Pandas

```python
#importando dados csv
# file_name = 'Sales2022.csv'
caminho_arquivo =  r'C:\Users\Maycon Rocha\My_Projects\Fundamentos_Python\Sales2022.csv'

data_tipe = {'CustomerNumber': str, 'SalesOrder':str}

df = pd.read_csv(caminho_arquivo, dtype= data_tipe, parse_dates=['SalesDate'])
```

### Selecionando Colunas


```python
df.columns

colunas = ['Country', 'State', 'City', 'PlantName', 'MaterialNumber', 'BusinessLine', 'Category', 'SalesDate',  'SalesOrder', 'Volume', 'Revenue']

df[colunas]

df[['MaterialNumber', 'Category']].head()
```

### Filtrando dados


```python
# df['PlantName'].unique()

sorocaba_df = df[df['PlantName']=='Sorocaba']
sorocaba_df.head()
```

```python
sorocaba_df[sorocaba_df['Revenue']>=1000].head()
```

```python
sorocaba_df[(sorocaba_df['SalesDate']>='2022-01-01') & (sorocaba_df['SalesDate']<='2022-01-31')].sort_values(by='SalesDate', ascending=False)
```

### Group by

```python
sorocaba_df.groupby('Category')['Revenue'].sum().sort_values(ascending=False)
```

```python
sorocaba_df.groupby(['BusinessLine', 'Category'])['Revenue'].sum().sort_values(ascending=False)
```

```python
sorocaba_df_group = sorocaba_df.groupby(['BusinessLine', 'Category'])['Revenue'].sum().sort_values(ascending=False).reset_index()
sorocaba_df_group
# type(sorocaba_df_group)
```

```python
type(sorocaba_df_group)
```

### Tabelas dinâmicas

```python
df.pivot_table(
    index=  'State',
    columns= 'BusinessLine',
    values='Revenue',
    aggfunc='sum',
    margins= True,
    margins_name= 'Total'
    
)
```
